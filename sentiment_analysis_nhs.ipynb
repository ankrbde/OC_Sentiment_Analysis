```python
import pandas as pd
import numpy as np
from google_play_scraper import app, Sort, reviews_all
import plotly.express as px
```

```python
nhs_project = reviews_all( 'com.nhs.online.nhsonline' , sleep_milliseconds=0, lang='en', country='GB', sort=Sort.NEWEST )
```

```python
#nhs_project
```

```python
df = pd.json_normalize(nhs_project)
```

```python
print(df.shape)
df = df.sort_values(by=['at'], ascending=False).head(1000)
print(df.shape)
```

```python
df.head()
```

```python
from transformers import pipeline

sentiment_analysis = pipeline("sentiment-analysis",model="siebert/sentiment-roberta-large-english", batch_size =8, truncation=True)
```

```python
df['content'] = df['content'].astype('str')
```

```python
df['result'] = df['content'].apply(lambda x: sentiment_analysis(x))
```

```python
df.head()
```

```python
df['sentiment'] = df['result'].apply(lambda x: x[0]['label'])
df['resultScore'] = df['result'].apply(lambda x: x[0]['score'])
```

```python
df.head()
```

```python
df['sentiment'].value_counts(normalize=True)
```

```python
fig = px.histogram(df, x='sentiment', color='sentiment', text_auto=True)
fig.show()
```
